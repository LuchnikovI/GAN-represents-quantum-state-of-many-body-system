### Here we generate and save samples from TFI model

```
#spin chain length
N = 32

#magnetic field along x axis
ta = 0.9

#TFI model
model = ising.ising_chain(N, ta, la=0.00001)
model.gs()

#TFI model saving
filehandler = open("TFI data/model_N=" + str(N) + "_h=" + str(round(ta, 4)) + ".obj", 'wb') 
pickle.dump(model, filehandler)


#10^6 samples
samples = np.array([model.sample(100000) for _ in range(10)]).reshape((1000000, -1))

#saving
np.save("TFI data/samples_N=" + str(N) + "_h=" + str(round(ta, 4)) + ".npy", samples)
```
